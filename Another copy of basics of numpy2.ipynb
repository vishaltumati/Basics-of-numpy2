```python
# Creating random matrices

#between 0 and 1
rand1 = np.random.rand(3,3)
print(rand1)

# between a and b
rand2 = np.random.randint(10,20,(3,3))
print(rand2)
```

```python
# Array properties

arr = np.array([[1,2,3,4,],[0,3,4,5,]])
print(arr)

# shape of array
print(f' the shape of the array is {arr.shape}')

# size of array
print(f' the size of the array is {arr.size}')

#Datatype of array
print(f' the Data type of the array is {arr.dtype}')

# Dimensions of array
print(f' the dimension of the array is {arr.ndim}')
```

```python
# Indexing and sliding
# Access the elements in 1D array

arr = np.array([1,2,3,4])
print(arr)
print(arr[3])


# sliding the elements
print(arr[0:3])

#Access the elements in 2D array
arr2 = np.array([[1,2,3],[2,3,4]])
print(arr2)
print(arr2[1,1])
print(arr2[0,:]) # Frist row with all columns
print(arr2[:,0]) #Frist columns with all column
```

```python
# l1 = []
# [l1.append(_) for _ in range(10)]

l1 = np.arange(10,-10,-3)
print(l1)
```

```python
# Mathematics operations

a = np.array([1,2,3])
b = np.array([3,4,5])

print(f' the sum of a and b is {a + b}')
print(f' the difference of a and b is {a - b}')
print(f' the product of a and b is {a * b}')
print(f' the division of a and b is {a / b}')
```

```python
# Matrix multiplication
b1 = np.array([[1,2,3],[2,3,4]])
b2 = np.array([[5,5],[8,8],[1,1]])

# Dot product --- product = AB -- A m x n -- B n x p
c = np.dot(b1,b2)
print(c)
```
